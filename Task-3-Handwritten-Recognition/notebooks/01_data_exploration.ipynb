# Handwritten Character Recognition — Data Exploration

A short look at the MNIST data and the augmentation pipeline used to train the CNN.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

import numpy as np
import matplotlib.pyplot as plt

from src.utils import load_config
from src.dataset import DataLoader

cfg = load_config('../config.yaml')
data = DataLoader(cfg).load()
print('train', data.x_train.shape, 'val', data.x_val.shape, 'test', data.x_test.shape)
```

## Sample digits

```python
fig, axes = plt.subplots(3, 8, figsize=(12, 5))
for ax, i in zip(axes.ravel(), np.random.choice(len(data.x_train), 24, replace=False)):
    ax.imshow(data.x_train[i].squeeze(), cmap='gray')
    ax.set_title(int(data.y_train[i]))
    ax.axis('off')
plt.tight_layout(); plt.show()
```

## Class balance

```python
vals, counts = np.unique(data.y_train, return_counts=True)
plt.bar(vals, counts)
plt.xlabel('Digit'); plt.ylabel('Count'); plt.title('Training class distribution')
plt.xticks(vals); plt.show()
```

## Augmentation preview
The same digit transformed by the in-graph augmentation layers.

```python
from src.model import CNNModel
aug = CNNModel(cfg)._augmentation()
sample = data.x_train[0:1]
fig, axes = plt.subplots(1, 6, figsize=(12, 2.5))
axes[0].imshow(sample.squeeze(), cmap='gray'); axes[0].set_title('original'); axes[0].axis('off')
for ax in axes[1:]:
    ax.imshow(aug(sample, training=True).numpy().squeeze(), cmap='gray')
    ax.set_title('augmented'); ax.axis('off')
plt.tight_layout(); plt.show()
```
